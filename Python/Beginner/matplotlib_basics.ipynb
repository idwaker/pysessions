```python
x=[[6],[8],[10],[14],[18]]
y=[[7],[9],[13],[17.5],[18]]

plt.plot(x,y,'.')
plt.title("Diameter of pizza against price")
plt.xlabel("Diameter in inches")
plt.ylabel("Price in Dollar")
plt.show()
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
model.predict(20)[0]
```

```python
%matplotlib inline
from random import randint
a = [randint(0,10) for _ in range(10)]
b = [randint(0,10) for _ in range(10)]
plt.plot(a,b,'^',color='red')
plt.xlabel('random numbers from a')
plt.ylabel('random numbers from b')
plt.title('Just some random numbers')
```

```python
aa = list(range(10))
bb = [randint(1,10) for _ in range(10)]
print(aa)
print(bb)
plt.bar(aa,bb,label="random numbers")
plt.legend(loc='best')
```
