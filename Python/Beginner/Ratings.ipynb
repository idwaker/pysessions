```python
import csv
import numpy as np
from math import sqrt

f = open(r'/media/mohit/mohit1/githubfile/dataset/ratings.csv','r')
data = np.array(list(csv.reader(f)))

data2 = np.array(data[1:,1:],dtype=np.float)
print(data2)

def dist(dataa,datab):
    d = (dataa - datab)**2
    return (d.sum())**0.5
        
def mmain():
    a=[]
    for i in range(len(data2)):
        aa = []
        for j in range(len(data2)):
                s=dist(data2[i],data2[j])
                aa.append(s)
        a.append(aa)
    print("-----------------------")
    print(np.array(a))
    print("-----------------------")
    print(np.array(a).mean())
    print("-----------------------")
    b=0
    c=0
    for i in range(len(a)):
        for j in range(len(a)):
            if i!=j:
                c+=1
                b+=a[i][j]
    avg=b/c
    print(avg)
    for i in range(len(a)):
        for j in range(len(a)):
                if a[i][j]<=avg:
                    a[i][j]=1
                else:
                    a[i][j]=(0)
    print(np.array(a))   

mmain()
    
```
